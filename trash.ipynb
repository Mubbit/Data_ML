```python
import random

next_id=pd.DataFrame([random.randint(100000,999999)]*3,columns=['next_id'])
next_id.shape
```

```python
sample_case_list_wni=pd.concat([sample_case_list,next_id],axis=1)
sample_case_list_wni
```

```python
import re
summarys=[]
jmreferences=[]
for i in range(3):
    summary=re.sub('법','<highlighted>법</highlighted>',sample_case_list.loc[i,'summary'])
    jmreference=re.sub('법','<highlighted>법</highlighted>',sample_case_list.loc[i,'jmreference'])
    
    summarys.append(summary)
    jmreferences.append(jmreference)
    
summ_jm=pd.concat([pd.DataFrame(summarys,columns=["summary_textrank"]),pd.DataFrame(jmreferences,columns=['jmreference_textrank'])],axis=1)
summ_jm
```

```python
sample_case_list_wtr=pd.concat([sample_case_list_wni,summ_jm],axis=1)
sample_case_list_wtr
```

```python
kobart=[]
for i in range(3):
    kobart.append(sample_case_list.loc[i,'jmreference']+sample_case_list.loc[i,'summary'])
sample_case_list_wkb=pd.concat([sample_case_list_wtr,pd.DataFrame(kobart,columns=['kobart']),pd.DataFrame(kobart,columns=['_'])],axis=1)
sample_case_list_wkb
```

```python
sample_case_list_wkb['datanumber']=sample_case_list_wkb['datanumber'].apply(int)
sample_case_list_wkb['next_id']=sample_case_list_wkb['next_id'].apply(int)
sample_case_list_wkb.to_csv('sample_case_list.csv',index=False)
```
